```python
# https://wikidocs.net/137791
def draw(style,grid=False):
    plt.title('나이와 자산의 상관관계')
    plt.plot(range(10,50,10),[1,4,9,16],style)
    plt.xlabel('나이')
    plt.ylabel('재산')
    plt.grid(grid) # 그리드 설정
    plt.show()

draw('bo--',grid=True)
```

### 여러 개의 선 그어보기

```python
val = np.arange(0,5,0.2)
val
```

```python
# 1. 여러 개의 선 그리기 (하나의 plot 사용)
plt.title('라인 플롯에 여러개 선')
plt.plot(
    val,val,'r--',
    val,val**2,'bo-',
    val,np.sin(val**2),'g*-'
)
plt.show()
```

```python
# 2. 하나의 plot이 아니라 여러 개 -> 겹치는.
# 여러개의 plot을 사용하여 겹치게
plt.plot(np.arange(4+1)**2)
plt.show()
plt.plot(np.arange(4-1)**3)
plt.show()
```

```python
# 여러 개의 플롯이 겹쳤을 경우 라벨 표시
x = np.linspace(-np.pi,np.pi,256)
```

```python
# 코사인, 사인 곡선
c,s = np.cos(x),np.sin(x)
plt.plot(x,c,label='Cosine')
plt.plot(x,s,label='Sin')
plt.legend()
plt.show()
```

### 다양한 차트 활용
* `lineplot` 선 차트
* 그외에 여러 가지 차트가 존재

```python
y = [2,3,1]
x = np.arange(3)
xlabel = ['가','나','다']
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.bar(x,y)
plt.show()
```

#### 파이 차트
* 원형의 파이 모양 그래프

```python
labels = ['라즈베리파이','딸기파이','토마토파이']
size=[15,30,45]
colors = ['yellowgreen','gold','lightskyblue']
explode = [0,0.1,0.1]
plt.title('pi chart')
plt.pie(size,explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%',shadow=True,startangle=90)
plt.axis('equal')
plt.show()
```

#### 히스토그램 (Histogram)

```python
x = np.random.randn(100)
plt.figure(figsize=(15,5))
plt.title('histogram')
plt.hist(x,bins=20)
plt.show()
```


#### 스캐터 플롯 (산점도)

* x과 y 값 사이의 상관성이 있는 변수들
* 두 변수 사이의 상관관계를 점들의 밀집 정도를 통해서 확인하는 그래프

```python
x = np.random.normal(0,1,100)
y = np.random.choice([0,1],size=100)
plt.title('scatter')
plt.scatter(x,y)
plt.show()
```

### 플롯 종류 선택 (`kind`)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

iris : pd.DataFrame = sns.load_dataset('iris') # 붓꽃
titanic : pd.DataFrame = sns.load_dataset('titanic') # 타이타닉호 데이터
```

```python
iris.columns
```

```python
iris['sepal_length'][:20].plot()
```

```python
# kind = 종류 => 해당 종류 plot 교체
iris['sepal_length'][:20].plot(kind='bar')
plt.title('꽃받침 길이 시각화')
plt.xlabel('data')
plt.ylabel('길이')
plt.show()
```

```python
# kind 를 사용할 필요 없이 plot 다음에 직접적으로 해당 차트의 메소드 호출해서 사용
iris[:5].plot.bar(rot=0)
plt.ylim(0,7)
plt.show()
```

```python
# 파이 차트 : 원형, 비율을 나타낼 때
df2 = titanic['class'].value_counts()
df2.plot.pie(autopct='%.2f%%')
plt.show()
```

```python
# 히스토그램
# - 가로축에 분포(값), 세로축에 빈도(몇개...)를 시각화
iris.plot.hist()
```

```python
# 커널 밀도 그래프
# - 부드럽게 그린 히스토그램
iris.plot.kde()
```

```python
# 박스플롯
# 5대 요약 수치 제공 + 이상치 추론
iris.boxplot(by='species')
plt.tight_layout(pad=3,h_pad=1)
plt.title('각 feature에 대한 boxplot')
plt.show()
```

```python
# 산점도 : 두 데이터의 관계를 점으로 표현
iris.plot.scatter('petal_length','petal_width')
```
